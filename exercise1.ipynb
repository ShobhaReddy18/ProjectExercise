Importing NumPy

```python
import numpy as np  # Import NumPy with an alias for easy usage
```

Creating a NumPy Array

```python
arr = np.array([1, 2, 3, 4, 5])
print("NumPy Array:", arr)
```

Basic Operations with NumPy

```python
# Creating an array
arr = np.array([10, 20, 30, 40, 50])
```

```python
# Performing basic arithmetic
print("Array + 5:", arr + 5)  # Adds 5 to each element
print("Array * 2:", arr * 2)  # Multiplies each element by 2
print("Mean of array:", np.mean(arr))  # Calculates mean
print("Standard Deviation:", np.std(arr))  # Calculates standard deviation
```

5. Selecting Specific Columns in a DataFrame

Selecting a Single Column

```python
#Selecting Columns:
#Use df['column'] to select data.
a['BatteryPower']       # Selecting the 'BatteryPower ' column
```

Selecting Multiple Columns

```python
#Use df[['col1', 'col2']] to select multiple columns.
a[['ScreenWidth', 'PriceRange']]
```

Selecting Multiple Columns for a Specific Row Range

```python
#Selecting multiple columns for the rows from 1000 to 2000
a[['BatteryPower', '4G', 'PriceRange']][1000:2000]
```

```python
# Selecting 'BatteryPower', '4G', and 'PriceRange' columns for every alternate row (step=2) within the positional index range from 1000 to 2000
a[['BatteryPower', '4G', 'PriceRange']][1000:2000:2]
```

```python
a[['BatteryPower', '4G', 'PriceRange']][0:1000:2000]  # Selecting with a step of 2000
```
